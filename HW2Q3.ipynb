```python
import numpy as np
import sympy as sym
import networkx as nx
import matplotlib.pyplot as plt
from dimod import BinaryQuadraticModel, ExactSolver
```

```python
el_num = 3
w = np.array([10, 20, 30])
v = np.array([60, 100, 120])
```

```python
max_w = 50
```

```python
y = np.empty(max_w, dtype=object)
for i in range(max_w):
    y[i] = sym.Symbol(f"y_{i+1}",bool=True)
```

```python
x = np.empty(el_num, dtype=object)
for i in range(el_num):
    x[i] = sym.Symbol(f"x_{i+1}",bool=True)
```

```python
f_1 = sym.expand((1-np.sum(y[:]))**2)
```

```python
f_1
```

```python
f_2_1 = 0
for i in range (max_w):
    f_2_1 = f_2_1 + (i+1)*y[i]
```

```python
f_2_1
```

```python
f_2_2 = 0
for i in range (el_num):
    f_2_2 = f_2_2 + w[i]*x[i]
```

```python
f_2 = sym.expand((f_2_1 - f_2_2)**2)
```

```python
f_2
```

```python
f_3 = 0
for i in range (el_num):
    f_3 = f_3 + v[i]*x[i]
```

```python
A = 10
B = 4
```

```python
f_t = A*(f_1 + f_2) - B*(f_3)
```

```python
f_t
```

```python
Q = np.zeros((el_num+max_w, el_num+max_w), dtype=float)
```

```python
for i in range(el_num+max_w):
    for j in range(i, el_num+max_w):
        if i<el_num and j<el_num:
            Q[i, j] = float(f_t.coeff(x[i]*x[j]))
        elif i<el_num and j>=el_num:
            Q[i, j] = float(f_t.coeff(x[i]*y[j-el_num]))
        elif i>=el_num and j>=el_num:
            Q[i, j] = float(f_t.coeff(y[i-el_num]*y[j-el_num]))
            
```

```python
bqm       = BinaryQuadraticModel.from_qubo(Q)
sampler   = ExactSolver()
sampleset = sampler.sample(bqm)
result    = sampleset.first
```
